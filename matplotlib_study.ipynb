```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 생성
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Figure와 Subplot 생성
fig, axs = plt.subplots(1, 3, figsize=(8, 3))

# 첫 번째 그래프
axs[0].plot(x, y1, label='sin(x)')
axs[0].set_title('Sine Function')
axs[0].legend()

# 두 번째 그래프
axs[1].plot(x, y2, label='cos(x)', color='orange')
axs[1].set_title('Cosine Function')
axs[1].legend()

# 세 번째 그래프
axs[2].plot(x, y3, label='tan(x)', color='green')
axs[2].set_title('Tangent Function')
axs[2].set_ylim(-10, 10)  # 탄젠트 함수의 범위 제한
axs[2].legend()

# 레이아웃 조정
plt.tight_layout()

# 그래프 표시
plt.show()
```

## np.random module 에서 자주 사용하는 methods

```python
x1 = np.random.normal(0,1,10000)
x2 = np.random.normal(-5,3,10000)
x3 = np.random.normal(5,2,10000)
d = dict(histtype='stepfilled', alpha=0.3, bins=50) # pack 

plt.hist(x1, **d) # unpack
plt.hist(x2, **d)
plt.hist(x3, **d);
```

```python
import numpy as np

# Create coordinate vectors
x = np.linspace(-5, 5, 10)
y = np.linspace(-5, 5, 10)

# Generate 2D grid coordinates
X, Y = np.meshgrid(x, y)

# X and Y are now matrices where each element corresponds to a point on the grid
print("X:\n", X)
print("Y:\n", Y)
```

```python
import matplotlib.pyplot as plt

# Define a function to evaluate
Z = np.sin(np.sqrt(X**2 + Y**2))

# Plot the function over the grid
plt.contourf(X, Y, Z, cmap='viridis')
plt.colorbar()
plt.title('Contour Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

## matplotlib subplot 그리는 방법

2. plt.subplots()

3. GridSpec
    - import matplotlib.gridspec as gridspec

```python
# plt.subplots()
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)

# First subplot
plt.subplot(211)  # (rows, cols, index)
plt.plot(x, np.sin(x))
plt.title('Sine Wave')

# Second subplot
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
plt.title('Cosine Wave')

plt.tight_layout()
plt.show()
```

```python
# plt.subplots()
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Top-left subplot
axs[0, 0].plot(x, np.sin(x))
axs[0, 0].set_title('Sine Wave')

# Top-right subplot
axs[0, 1].plot(x, np.cos(x))
axs[0, 1].set_title('Cosine Wave')

# Bottom-left subplot
axs[1, 0].plot(x, np.tan(x))
axs[1, 0].set_title('Tangent Wave')
axs[1, 0].set_ylim(-10, 10)  # Limit y-axis for tangent

# Bottom-right subplot
axs[1, 1].plot(x, -np.sin(x))
axs[1, 1].set_title('Negative Sine Wave')

plt.tight_layout()
plt.show()
```

```python
# GridSpec
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(8, 6))
gs = gridspec.GridSpec(3, 3)

ax1 = fig.add_subplot(gs[0, :])
ax2 = fig.add_subplot(gs[1:, :-1])
ax3 = fig.add_subplot(gs[1:, -1])

ax1.plot(x, np.sin(x))
ax2.plot(x, np.cos(x))
ax3.plot(x, np.tan(x))

plt.tight_layout()
plt.show()
```

## pandas 로 시각화. DataFrame.plot()
- plot(kind='line) -> plot(x)
- plot(kind='bar') -> bar(x, height)
- plot(kind='hist') -> hist(x, bins)
- pandas.plot() 은 간결하게 사용 가능하지만 정교한 작업은 불가능 함
- 인자값 : x, y, kind, subplots, figsize, title, style, xticks, yticks, xlim, ylim, xlabel, ylabel
