```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
# scatter plot 
sns.relplot(x="bill_length_mm", y="bill_depth_mm", data=penguins);
```

```python
# hue 사용하기
sns.relplot(x="bill_length_mm", y="bill_depth_mm", hue="bill_length_mm", data=penguins);
```

```python
sns.relplot(x="bill_length_mm", y="body_mass_g", hue="species", style="species", data=penguins);
```

```python
# col 을 사용하여 col 별로 여러 그래프 그리기
sns.relplot(x="bill_length_mm", y="body_mass_g", hue="species", col="island", data=penguins);
```

```python
# line plot
flights = sns.load_dataset("flights")
flights.head()
```

```python
# add 'line' in kind to draw line graph
# set 'kind' to draw 'line'
sns.relplot(x="year", y="passengers", kind="line", data=flights); # area is 95% confidence interval
```

```python
# set 'estimator' to 'None'
sns.relplot(x="year", y="passengers", kind="line", estimator=None, data=flights); # area is 95% confidence interval
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
tips = sns.load_dataset("tips")

# Create a relational plot
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="day",
    size="size",
    style="smoker",
    kind="scatter",
    col="time",
    height=4,
    aspect=1
)

plt.show()
```

```python
# date dataframe 생성
tdf = pd.DataFrame(np.random.randn(40,4),
                   index=pd.date_range('2020-01-01', periods=40),
                   columns=['A','B','C','D'])
```

```python
# date 로 x 축 표현하기
g = sns.relplot(kind='line', data=tdf)
g.fig.autofmt_xdate()
```

```python
# 범주형 산점도 ( stripplot(), swarmplot() )
sns.catplot(x="species", y="body_mass_g", hue="species", data=penguins);
## jitter 제거
sns.catplot(x="species", y="body_mass_g", hue="species", jitter=False, data=penguins); # jitter 제거
## set kind to 'swarm'
sns.catplot(x="species", y="body_mass_g", hue="sex", kind="swarm", data=penguins);
## set order to change x-axis name
```

```python
# 번주형 분포도(Categorical distribution plots)
## boxplot(), boxenplot(), violinplot()
```

```python
sns.catplot(x="species", y="body_mass_g", hue="sex", kind="box", data=penguins);
# set 'dodge' to overlap
sns.catplot(x="species", y="body_mass_g", hue="sex", kind="box", dodge=False, data=penguins);
## x, y 를 바꿔서 옆으로, 이쁘게 그릴 수 있음, 
## height, aspect 설정으로 세로, 가로 길이 설정 가능
```

```python
# x, y 설정 없이 자동으로 그리기
## orient 로 가로로 그리기 가능
sns.catplot(kind="box", orient='h', data=penguins)
```
